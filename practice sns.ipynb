```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
```

```python
df
```

```python
sns.relplot(x= 'year', y = 'passengers', data = df)
plt.show()
sns.set()
```

```python
sns.relplot(x= 'year', y = 'passengers', data = df, kind = 'line')
plt.show()
```

```python
sns.lineplot(x= 'year', y = 'passengers', data = df)
plt.show()
```

```python
sns.relplot(x= 'year', y = 'passengers', data = df, kind = 'line', hue = "month")
plt.show()
```

```python
sns.relplot(x= 'year', y = 'passengers', data = df, kind = 'line', hue = "month", col="month")
plt.show()
```

```python
sns.catplot(x= 'year', y = 'passengers', data = df)
plt.show()
```

```python
sns.swarmplot(x= 'year', y = 'passengers', data = df  )
plt.show()
# this does not allow overlapping of the points so it is easier to visualise
```

```python
sns.catplot(x= 'year', y = 'passengers', data = df, jitter = 0.002)
# this reduses the randomness of the values
plt.show()
```

```python
sns.catplot(x= 'year', y = 'passengers', data = df,kind ="swarm" )
plt.show()
# this does not allow overlapping of the points so it is easier to visualise
```

```python
sns.catplot(x= 'year', y = 'passengers', data = df,kind ="box" )
plt.show()
```

```python
sns.catplot(x= 'year', y = 'passengers', data = df,kind ="boxen")
plt.show()
```

```python
sns.catplot(x= 'year', y = 'passengers', data = df,kind ="violin")
plt.show()
```
